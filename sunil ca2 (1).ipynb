```python
#boxplot
sns.boxplot(data.rating)
plt.show()
```

```python
#Histoplot
sns.histplot(data.counts_of_ingredients)
plt.show()
```

```python
#rugplot
sns.rugplot(data.fourth_taste)
plt.show()
```

```python
#vilionplot
sns.violinplot(data.counts_of_ingredients)
plt.show()
```

```python
#stripplot
sns.stripplot(data.rating)
plt.show()
```

```python
#  PIE CHART
df = data['rating'].value_counts()
plt.pie(df, labels=df.index, autopct="%.0f%%");
```

```python
# NUMERICAL - NUMERICAL ASSOCIATIONS
plt.scatter(data.company_location,data.rating)
plt.show()
```

```python
# Correlation Matrix
sns.heatmap(data[["counts_of_ingredients","rating"]].corr(), annot=True, cmap="Blues")
```

```python
# Violin Plot
sns.violinplot(x=data["cocoa_percent"], y=data["counts_of_ingredients"])
```

```python
#heatmap
plt.subplots(figsize=(29,8))
corr = data.corr()
sns.heatmap(corr,annot=True,
           xticklabels = corr.columns.values,
           yticklabels = corr.columns.values);
plt.show()
```

```python
#pairplot
sns.pairplot(data,hue="company_location",size=3)
```
